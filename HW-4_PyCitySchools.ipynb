```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(20)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
grouped_schools_df = school_data_complete.groupby(['school_name'])
total_schools = len(grouped_schools_df["school_name"])
#print(f' Total number of schools: {total_schools}')

#Calculate the total number of students
total_students = school_data_complete["student_name"].count()
#print(f' Total number of students: {total_students}')

#Calculate the total budget
total_budget = grouped_schools_df['budget'].unique().sum()
#print(f' Total Budget: {float(total_budget)}')

#Calculate the average math score
Average_math_score = school_data_complete["math_score"].mean()
#print(f' Average Math score: {Average_math_score}')

#Calculate the average reading score
Average_reading_score = school_data_complete["reading_score"].mean()
#print(f' Average Reading score: {Average_reading_score}')

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
Passing_rate = (Average_math_score + Average_reading_score)/2
#print(f' Overall passing rate: {Passing_rate}')

#Calculate the percentage of students with a passing math score (70 or greater)
Math_score_ge_70 = school_data_complete.query('math_score>=70')
Percent_math_ge_70 = 100*Math_score_ge_70["student_name"].count()/school_data_complete["student_name"].count()
#print(f' Percent with math score 70 or more: {Percent_math_ge_70}')

#Calculate the percentage of students with a passing reading score (70 or greater)
Read_score_ge_70 = school_data_complete.query('reading_score>=70')
Percent_read_ge_70 = 100*Read_score_ge_70["student_name"].count()/school_data_complete["student_name"].count()
#print(f' Percent with reading score 70 or more: {Percent_read_ge_70}')

#Create a dataframe to hold the above results
results_data = {'Schools': [total_schools],
               'Students': [total_students],
               'Total Budget': ['${:,.2f}'.format(float(total_budget))],
               'Average Math Score': [Average_math_score],
               'Average Reading Score': [Average_reading_score],
               '% Passing Math': [Percent_math_ge_70],
               '% Passing Reading': [Percent_read_ge_70],
               '% Ovrall Passing Rate': [Passing_rate]} 
results_df = pd.DataFrame(results_data, index = [0])
results_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# Initializing lists for storing results 
Schools = []
School_type = []
Students = []
Total_budget_school = []
Student_budget = []
Average_math_score = []
Average_reading_score = []
Passing_math = []
Passing_reading = []
Overall_passing = []

# Collecting school names
school_names = school_data_complete["school_name"].unique()

# Looping over school names
for school in school_names:
    mask = school_data_complete["school_name"].isin([school])
    masked_df = school_data_complete[mask]

    Schools.append(school)

# Total Students
    total_students = int(masked_df.count().unique())

    Students.append(total_students)

# School Type
    school_type = masked_df["type"].unique()
#    print(f' School: {school}, School Type: {school_type[0]}, Total Students: {total_students}')

    School_type.append(school_type[0])

# Total budget
    total_school_budget = masked_df['budget'].unique().sum()
#    print(f' Total Budget for school {school}: {float(total_school_budget)}')
    Total_budget_school.append('${:,.2f}'.format(total_school_budget))

# Per student budget
    Budget_per_student = float(total_school_budget/total_students)
#    print(f' Per student budget for school {school}: {float(Budget_per_student)}')
    
    Student_budget.append('${:,.2f}'.format(Budget_per_student))

# Average math score per school
    Average_math_score_ps = masked_df['math_score'].mean()
#    print(f' Average Math score for school {school}: {Average_math_score_ps}')

    Average_math_score.append(Average_math_score_ps)

# Average reading score per school
    Average_reading_score_ps = masked_df['reading_score'].mean()
#    print(f' Average Reading score for school {school}: {Average_reading_score_ps}')
    
    Average_reading_score.append(Average_reading_score_ps)
    
# Percentage of students with a passing math score (70 or greater)
    Math_score_ge_70_ps = masked_df.query('math_score>=70')
    Percent_math_ge_70_ps = 100*Math_score_ge_70_ps['student_name'].count()/total_students
#    print(f' Percent with math score 70 or more in school {school}: {Percent_math_ge_70_ps}')

    Passing_math.append(Percent_math_ge_70_ps)
    
# Percentage of students with a passing reading score (70 or greater)
    Reading_score_ge_70_ps = masked_df.query('reading_score>=70')
    Percent_reading_ge_70_ps = 100*Reading_score_ge_70_ps['student_name'].count()/total_students
#    print(f' Percent with reading score 70 or more in school {school}: {Percent_reading_ge_70_ps}')

    Passing_reading.append(Percent_reading_ge_70_ps)
    
# Overall passing rate (math+reading)/2

    Overall_passing_rate_ps = (Percent_math_ge_70_ps+Percent_reading_ge_70_ps)/2
#    print(f' Overall passing rate in school {school}: {Overall_passing_rate_ps}\n')
    
    Overall_passing.append(Overall_passing_rate_ps)

#Create an overview table  
overview_table_results = {'School name': Schools,
                          'School Type': School_type,
                          'Students': Students,
                          'School Budget': Total_budget_school,
                          'Student Budget': Student_budget,
                          'Average Math Score': Average_math_score,
                          'Average Reading Score': Average_reading_score,
                          '% Passing Math': Passing_math,
                          '% Passing Reading': Passing_reading,
                          '% Overall Passing Rate': Overall_passing}


overview_table_results_df = pd.DataFrame(overview_table_results)
#overview_table_results_df.head(15)
Top_schools = overview_table_results_df.sort_values(by=['% Overall Passing Rate'], ascending=False)
Top_schools.head()
```

```python
# Note: This is the Math Score analysis result, next is the Reading Score analysis
# Initializing lists for storing results 
Grade_9 = []
Grade_10 = []
Grade_11 = []
Grade_12 = []

All_grades = []

Average_math_score_9th = []
Average_math_score_10th = []
Average_math_score_11th = []
Average_math_score_12th = []

# Collecting school names
school_grade = school_data_complete["grade"].unique()
#school_grade = ['9th', '10th', '11th', '12th']
school_names = school_data_complete["school_name"].unique()
#print(school_grade)
# Looping over school names
for grade in school_grade:
    
    if grade == '9th':
        mask_grade = school_data_complete["grade"].isin([grade])
        masked_grade_df = school_data_complete[mask_grade]
        data_grades_rscores_9th = masked_grade_df[['school_name', 'grade', 'math_score']]

        Grade_9.append(masked_grade_df)      
#        print(grade)
        
        for school in school_names:
            mask = data_grades_rscores_9th["school_name"].isin([school])
            masked_df_9th = data_grades_rscores_9th[mask]
            score = masked_df_9th['math_score'].mean()
            # print(f' Average Reading score for school {school}: {Average_reading_score_ps}')
            Average_math_score_9th.append(score)
#            print(f' School: {school} has average reading score: {score}')
    if grade == '10th':
        mask_grade = school_data_complete["grade"].isin([grade])
        masked_grade_df = school_data_complete[mask_grade]
        data_grades_rscores_10th = masked_grade_df[['school_name', 'grade', 'math_score']]
        
        Grade_10.append(masked_grade_df)
#        print(grade)
        for school in school_names:
            mask = data_grades_rscores_10th["school_name"].isin([school])
            masked_df_10th = data_grades_rscores_10th[mask]
            score = masked_df_10th['math_score'].mean()
            Average_math_score_10th.append(score)
    if grade == '11th':
        mask_grade = school_data_complete["grade"].isin([grade])
        masked_grade_df = school_data_complete[mask_grade]
        data_grades_rscores_11th = masked_grade_df[['school_name', 'grade', 'math_score']]
        
        Grade_11.append(masked_grade_df)
#        print(grade)
        for school in school_names:
            mask = data_grades_rscores_11th["school_name"].isin([school])
            masked_df_11th = data_grades_rscores_11th[mask]
            score = masked_df_11th['math_score'].mean()
            Average_math_score_11th.append(score)
    if grade == '12th':
        mask_grade = school_data_complete["grade"].isin([grade])
        masked_grade_df = school_data_complete[mask_grade]
        data_grades_rscores_12th = masked_grade_df[['school_name', 'grade', 'math_score']]
        
        Grade_12.append(masked_grade_df)
#        print(grade)

        for school in school_names:
            mask = data_grades_rscores_12th["school_name"].isin([school])
            masked_df_12th = data_grades_rscores_12th[mask]
            score = masked_df_12th['math_score'].mean()
            Average_math_score_12th.append(score)
    
All_grades.append(Grade_9)
All_grades.append(Grade_10)
All_grades.append(Grade_11)
All_grades.append(Grade_12)

d9 = dict(zip(school_names,Average_math_score_9th))
d10 = dict(zip(school_names,Average_math_score_10th))
d11 = dict(zip(school_names,Average_math_score_11th))
d12 = dict(zip(school_names,Average_math_score_12th))

d9_df = pd.DataFrame.from_dict(d9, orient='index', columns=['9th grade'])
d10_df = pd.DataFrame.from_dict(d10, orient='index', columns=['10th grade'])
d11_df = pd.DataFrame.from_dict(d11, orient='index', columns=['11th grade'])
d12_df = pd.DataFrame.from_dict(d12, orient='index', columns=['12th grade'])

merge1 = pd.merge(d9_df, d10_df, left_index=True, right_index=True)
merge2 = pd.merge(merge1, d11_df, left_index=True, right_index=True)
merge3 = pd.merge(merge2, d12_df, left_index=True, right_index=True)

final_data_bygrade = merge3
final_data_bygrade
```
